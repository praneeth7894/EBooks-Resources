```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#non stationary
df1=pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates= True)
df1.index.freq='MS'

#stationary
df2=pd.read_csv('DailyTotalFemaleBirths.csv', index_col='Date', parse_dates=True)
df2.index.freq='D'
```

```python
#pip install pmdarima --user
```

```python
from pmdarima import auto_arima
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
help(auto_arima)
```

```python
stepwisefit= auto_arima(df2['Births'], start_p=0, start_q=0, max_p=6, max_q=3, seasonal=False, trace=True)
```

```python
stepwisefit.summary()
```

```python
#the above model suggests us which model to use
```

```python
stepwisefit_s= auto_arima(df1['Thousands of Passengers'], trace=True, seaonsal=True, start_p=0, start_q=0, m=12)
#we need to mention m because it tells us how many seasons. Since there are 12 months, we put 12
```

```python
stepwisefit_s.summary()
```
