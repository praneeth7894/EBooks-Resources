# SARIMA

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
from pmdarima import auto_arima
```

```python
df1=pd.read_csv('co2_mm_mlo.csv', index_col='Date', parse_dates=True)
df1.head()
```

```python
df1.index.freq='MS'
```

```python
df1['interpolated'].plot(figsize=(12,8))
```

```python
ets=seasonal_decompose(df1['interpolated'], model='add')
ets.plot()
```

```python
auto_arima(df1['interpolated'], trace=True, m=12, seasonal=True).summary()
```

```python
train=df1.iloc[:717]
test=df1.iloc[717:]
```

```python
model=SARIMAX(train['interpolated'], order=(0,1,1), seasonal_order=(1,0,1,12)).fit()
model.summary()
```

```python
start=len(train)
end=len(train)+len(test)-1
```

```python
pred1=model.predict(start, end, typ='levels')
```

```python
pred1.plot(legend=True, label='pred')
test['interpolated'].plot(legend=True, label='test')
```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
rmse(pred1, test['interpolated'])
```

```python
test['interpolated'].mean()
```

```python
#forecast into unknown future
```

```python
model_f=model=SARIMAX(df1['interpolated'], order=(0,1,1), seasonal_order=(1,0,1,12)).fit()
model_f.summary()
```

```python
fcast=model_f.predict(len(df1), len(df1)+11, typ='levels')
```

```python
df1['interpolated'].plot(legend=True, label='train')
fcast.plot(legend=True, label='future')
```

# SARIMAX without Exogenous Variable

```python
df2=pd.read_csv('RestaurantVisitors.csv', index_col='date', parse_dates=True)
df2.index.freq='D'
```

```python
df2=df2.dropna()
```

```python
df2.head()
```

```python
df2.columns
```

```python
cols=['rest1', 'rest2', 'rest3','rest4', 'total']
```

```python
for i in cols:
    df2[i]=df2[i].astype(int)
```

```python
df2.head()
```

```python
ets2=seasonal_decompose(df2['total'])
ets2.plot();
```

```python
ets2.seasonal.plot(figsize=(18,5))
```

```python
train=df2.iloc[:436]
test=df2.iloc[436:]
```

```python
auto_arima(df2['total'], seasonal=True, m=7).summary()
```

```python
result2=SARIMAX(df2['total'], seasonal_order=(1,0,1,7), order=(0,0,0), enforce_invertibility= False).fit()
```

```python
result2.summary()
```

```python
start2=len(train)
end2=len(train)+len(test)-1
```

```python
final2=result2.predict(start2,end2, typ='levels')
final2.plot(legend=True, label='pred')
test['total'].plot(legend=True, label='actual')
```

```python
rmse(final2, test['total'])
```

```python
test['total'].mean()
```

```python
final2.mean()
```

```python
#we see a difference in plots here because we didnt add the exogenous variable ie holidays
```

# SARIMAX with Exogenous Variable

```python
auto_arima_ex=auto_arima(df2['total'], exogenous=df2[['holiday']], seasonal=True, m=7).summary()#double bracket because statsmodels exepects a dataframe
```

```python
auto_arima_ex
```

```python
model_ex=SARIMAX(train['total'], exog=train[['holiday']], order=(1,0,0), seasonal_order=(2,0,0,7), enforce_invertibility= False).fit()
```

```python
model_ex.summary()
```

```python
pred3=model_ex.predict(start2, end2, typ='levels', exog=test[['holiday']])
```

```python
pred3.plot(legend=True, label='Pred')
test['total'].plot(legend=True, label='actual', figsize=(12,8))
```

```python
rmse(test['total'], pred3)
```

```python
model_sarimax= SARIMAX(df2['total'], exog=df2[['holiday']], order=(1,0,0), seasonal_order=(2,0,0,7), enforce_invertibility= False).fit()
```

```python
df3=pd.read_csv('RestaurantVisitors.csv', index_col='date', parse_dates=True)
df3.index.freq='D'
```

```python
exog_forecast=df3[478:][['holiday']]#providing future holday dates
```

```python
pred_sarimax=model_sarimax.predict(len(df2), len(df2)+38, exog=exog_forecast)
```

```python
ax=df3['total'].loc['2017-01-01':].plot(legend=True, title='title')
for x in df2.query('holiday ==1').index:
    ax.axvline(x=x, color='k', alpha=0.8)
pred_sarimax.plot(legend=True, label='Future')
df2['total'].plot(legend=True, label='Past', figsize=(18,12))
```
